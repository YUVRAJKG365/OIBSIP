## Merge data (if necessary)

### Subtask:
Merge the two dataframes `df_india` and `df_upto_2020` if necessary for the analysis.


**Reasoning**:
Assess if merging is necessary based on the data exploration and clean dataframes. If needed, merge the two dataframes on common columns using an outer join to keep all records and then display the head and info of the merged dataframe.



```python
# Assess if merging is necessary
# df_india covers dates from 2019-05-31 to 2020-06-30
# df_upto_2020 covers dates from 2020-01-31 to 2020-10-31
# There is an overlapping period from 2020-01-31 to 2020-06-30.
# The columns related to unemployment rates and employment are common.
# df_upto_2020 also contains 'Region.1', 'longitude', and 'latitude' which are not in df_india.
# Merging is necessary to have a complete time series for analysis of unemployment rate and employed figures,
# and potentially use the geographical information from df_upto_2020.

# Identify common columns for merging
common_columns = ['Region', 'Date', 'Frequency', 'Estimated Unemployment Rate (%)', 'Estimated Employed', 'Estimated Labour Participation Rate (%)']

# Perform the merge operation using an outer join to keep all records from both dataframes
merged_df = pd.merge(df_india, df_upto_2020, on=common_columns, how='outer', suffixes=('_india', '_upto2020'))

# Display the head and information of the merged DataFrame
print("Head of merged_df:")
display(merged_df.head())

print("\nInfo of merged_df:")
merged_df.info()
```

## Analyze unemployment rate

### Subtask:
Analyze the unemployment rate over time, paying close attention to the period during COVID-19. Calculate relevant metrics and identify trends.


**Reasoning**:
The first step is to sort the merged_df by date in ascending order. Then, calculate the national average unemployment rate for each month by grouping by date and calculating the mean of 'Estimated Unemployment Rate (%)'. Next, calculate the average unemployment rate for each region over the entire time period by grouping by region and calculating the mean. Finally, identify the COVID-19 period, calculate the average unemployment rate before and during this period, and the percentage change between the two periods.



```python
# 1. Sort the merged_df DataFrame by the 'Date' column in ascending order.
merged_df = merged_df.sort_values(by='Date')

# 2. Calculate the national average unemployment rate for each month.
national_avg_unemployment_monthly = merged_df.groupby('Date')['Estimated Unemployment Rate (%)'].mean().reset_index()

# 3. Calculate the average unemployment rate for each region over the entire time period.
region_avg_unemployment = merged_df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().reset_index()

# 4. Identify the period corresponding to the COVID-19 pandemic (from March 2020 onwards).
# Assuming COVID-19 impact is significant from March 2020
covid_start_date = pd.to_datetime('2020-03-01')

# 5. Calculate the average unemployment rate during the identified COVID-19 period.
covid_period_df = merged_df[merged_df['Date'] >= covid_start_date]
avg_unemployment_during_covid = covid_period_df['Estimated Unemployment Rate (%)'].mean()

# 6. Calculate the average unemployment rate before the identified COVID-19 period.
before_covid_period_df = merged_df[merged_df['Date'] < covid_start_date]
avg_unemployment_before_covid = before_covid_period_df['Estimated Unemployment Rate (%)'].mean()

# 7. Calculate the percentage change in the average unemployment rate from before to during the COVID-19 period.
percentage_change = ((avg_unemployment_during_covid - avg_unemployment_before_covid) / avg_unemployment_before_covid) * 100

print("National Average Unemployment Rate (Monthly):")
display(national_avg_unemployment_monthly)

print("\nAverage Unemployment Rate by Region (Overall):")
display(region_avg_unemployment)

print(f"\nAverage Unemployment Rate Before COVID-19 ({before_covid_period_df['Date'].min().date()} to {before_covid_period_df['Date'].max().date()}): {avg_unemployment_before_covid:.2f}%")
print(f"Average Unemployment Rate During COVID-19 ({covid_period_df['Date'].min().date()} to {covid_period_df['Date'].max().date()}): {avg_unemployment_during_covid:.2f}%")
print(f"Percentage Change in Average Unemployment Rate (Before to During COVID-19): {percentage_change:.2f}%")
```

## Visualize data

### Subtask:
Create visualizations (e.g., line plots, bar charts) to illustrate the unemployment rate trends, regional differences, and the impact of COVID-19.


**Reasoning**:
Create the required visualizations to illustrate unemployment trends, regional differences, and the impact of COVID-19 using the previously calculated dataframes and values.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Line plot of national average unemployment rate over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=national_avg_unemployment_monthly, x='Date', y='Estimated Unemployment Rate (%)', marker='o', color='teal')
plt.title('National Average Unemployment Rate Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Estimated Unemployment Rate (%)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# 2. Bar chart of average unemployment rate by region
region_avg_unemployment_sorted = region_avg_unemployment.sort_values(by='Estimated Unemployment Rate (%)', ascending=False)
plt.figure(figsize=(15, 7))
sns.barplot(data=region_avg_unemployment_sorted, x='Region', y='Estimated Unemployment Rate (%)', palette='viridis', hue='Region', legend=False)
plt.title('Average Unemployment Rate by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Estimated Unemployment Rate (%)', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# 3. Bar chart comparing average unemployment rate before and during COVID-19
comparison_data = pd.DataFrame({
    'Period': ['Before COVID-19', 'During COVID-19'],
    'Average Unemployment Rate (%)': [avg_unemployment_before_covid, avg_unemployment_during_covid]
})

plt.figure(figsize=(8, 5))
sns.barplot(data=comparison_data, x='Period', y='Average Unemployment Rate (%)', palette='magma', hue='Period', legend=False)
plt.title('Average Unemployment Rate Before and During COVID-19', fontsize=16)
plt.xlabel('Period', fontsize=12)
plt.ylabel('Average Unemployment Rate (%)', fontsize=12)
plt.ylim(0, max(avg_unemployment_before_covid, avg_unemployment_during_covid) * 1.2) # Set y-limit for better visualization
plt.tight_layout()
plt.show()
```
